```python
from xml.etree import ElementTree as ET
import csv

# PARSE XML
xml = ET.parse("/Users/liuyifu/github_projects/xml_2_csv/data/doc.xml")

# CREATE CSV FILE
csvfile = open("data.csv",'w',encoding='utf-8')
csvfile_writer = csv.writer(csvfile)

# ADD THE HEADER TO CSV FILE
csvfile_writer.writerow(["name","role","age"])

# FOR EACH EMPLOYEE
for employee in xml.findall("SimpleField"):
    
    if(employee):
       
       # EXTRACT EMPLOYEE DETAILS  
      name = employee.find("name")
      role = employee.find("role")
      age = employee.find("age")
      csv_line = [name.text, role.text, age.text]

      # ADD A NEW ROW TO CSV FILE
      csvfile_writer.writerow(csv_line)
csvfile.close()  
```

```python
result = {}
# scope_keys = ["SimpleData", "coordinates"]
scope_keys = [
            "SimpleData", 
            "coordinates",
            'Placemark'
              ]
keys = set([elem.tag for elem in root.iter()])
for key in keys:
    key_txt = key.replace(r"{http://www.google.com/kml/ext/2.2}", "").replace("{http://www.opengis.net/kml/2.2}", "") 
    # print(key)
    if key_txt in scope_keys:

        for item in root.iter(key):
            cur_value = None
            if key_txt not in result:
                result[key_txt] = []
            
            if key_txt == 'Placemark':
                cur_value = item.attrib['id']

            elif key_txt == 'coordinates':
                cur_value = item.text
            
            result[key_txt].append(cur_value)
            
```
